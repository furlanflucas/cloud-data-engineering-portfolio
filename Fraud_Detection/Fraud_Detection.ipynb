```python
-- Total of flagged fraud by city 
SELECT 
	location, 
	COUNT(*)IsFlaggedFraud
FROM 'fraud_detection_dataset.csv'
	WHERE IsFlaggedFraud = 1 
GROUP BY location
ORDER BY IsFlaggedFraud DESC;



```

```python
-- Fraud by Device type
SELECT 
    Device,
    COUNT(*) AS FraudulentTransactions
FROM 'fraud_detection_dataset.csv'
WHERE IsFlaggedFraud = 1
GROUP BY Device
ORDER BY FraudulentTransactions DESC;
```

```python
-- Identifying Users with Suspicious Activity
SELECT UserID,
    COUNT(TransactionID) AS TotalTransactions,
    SUM(CASE WHEN IsFlaggedFraud = 1 THEN 1 ELSE 0 END) AS FraudulentTransactions,
    ROUND(SUM(CASE WHEN IsFlaggedFraud = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(TransactionID), 2) AS FraudRate,
    MAX(TransactionAmount) AS MaxTransactionAmount,
    MIN(TransactionDate) AS FirstTransactionDate,
    MAX(TransactionDate) AS LastTransactionDate
FROM 'fraud_detection_dataset.csv'
GROUP BY UserID
HAVING FraudulentTransactions > 3 
    OR FraudRate > 50
ORDER BY FraudulentTransactions DESC, FraudRate DESC;
```

```python
-- Locations with Frequent Policy Violations
SELECT 
    Location,
    COUNT(TransactionID) AS TotalTransactions,
    SUM(CASE WHEN IsFlaggedFraud = 1 THEN 1 ELSE 0 END) AS FraudulentTransactions,
    ROUND(SUM(CASE WHEN IsFlaggedFraud = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(TransactionID), 2) AS FraudRate
FROM 'fraud_detection_dataset.csv'
GROUP BY  Location
HAVING FraudRate > 10
ORDER BY FraudulentTransactions DESC;
```
